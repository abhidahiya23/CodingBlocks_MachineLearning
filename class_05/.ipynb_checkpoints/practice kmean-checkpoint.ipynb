```python
from sklearn.datasets import make_blobs
```

```python
X,_ = make_blobs(n_samples=1500, centers=6)
```

```python
from matplotlib import pyplot as plt
print(X.shape)
```

```python
plt.figure(0)
plt.scatter(X[:,0],X[:,1])
plt.show()
```

```python
import numpy as np
# defining clusters 
k = 6

colors = ['p','b','o','g','r','y']
clusters = {}

for ix in range(k):
        
        center = 12.0*( 2*np.random.random((X.shape[1],))-1 )
        
        points = []
        cluster = {  
            "cen" : center,
            "color" : colors[ix],
            "pts" : points
        }
        
        clusters[ix] = cluster
print(clusters)        
```

```python
for ix in range(k):
    plt.scatter(clusters[ix]['cen'][0],clusters[ix]['cen'][1],c = 'black',marker = "*",s = 1000)

plt.scatter(X[:,0],X[:,1]) 
plt.legend()
plt.show()
```
