```python
import numpy as np
from matplotlib import pyplot as plt
from sklearn.datasets import make_blobs
```

```python
X,_ = make_blobs(n_samples=1000 , centers=5 , random_state=1)
print(X.shape)
print(X.)
```

```python
#basic plot containing all clusters
plt.figure()
plt.scatter(X[:,0],X[:,1])
plt.show()
```

```python
#defining cluster limits

k = 5

colors = ['pink','blue','green','red','orange']

clusters = {}

for ix in range(k):
    center = 10.0*( 2*np.random.random((X.shape[1],))-1 )
    points = []
    
    cluster = {
        "center" : center,
        "points" : points,
        "color"  : colors[ix]

    }
    
    clusters[ix] = cluster
    
print(clusters)    
```

```python
#plotting clusters with their center
plt.figure()
plt.scatter(X[:,0],X[:,1])
for ix in range(k):
    plt.scatter(clusters[ix]["center"][0],clusters[ix]["center"][1],color = "black",marker="*",s=100)#s = size of cluster ceters
plt.show()    
```

```python
# distance function
def distance(v1,v2):
    return np.sqrt(np.sum((v1-v2)**2))
```

```python
# assigning points to array
for ix in range(X.shape[0]):
    
    dis = []
    curr_p = X[ix]
    
    for kx in range(k):
        d = distance(curr_p,clusters[kx]["center"])
        dis.append(d)
        
    assigned_cluster = np.argmin(dis)
    
    clusters[assigned_cluster]["points"].append(curr_p)

# np array
for kx in range(k):
    pnts = np.array(clusters[kx]["points"])
    
    clusters[kx]["coords"] = pnts
    
# figure 1
plt.figure(0)
for kx in range(k):
    pts = clusters[kx]["coords"]
    
    try:
        plt.scatter(pts[:,0],pts[:,1],color = clusters[kx]["color"])
    except:
        pass
    
    plt.scatter(clusters[kx]["center"][0],clusters[kx]["center"][1],color = "black",marker="*",s=100)

# updating centers
for kx in range(k):
    
    if clusters[kx]["coords"].shape[0]>0:
        new_center = clusters[kx]["coords"].mean(axis=0)
    else:
        new_center = clusters[kx]["center"]
        
    clusters[kx]["center"] = new_center
    clusters[kx]["points"] = []
    
# figure 1 with updated centers
plt.figure(1)
for kx in range(k):
    pts = clusters[kx]["coords"]
    
    try:
        plt.scatter(pts[:,0],pts[:,1], color = clusters[kx]["color"])
    except:
        pass
    
    plt.scatter(clusters[kx]["center"][0], clusters[kx]["center"][1], color = "black", marker = "*", s=100)
    
plt.show()    
```
