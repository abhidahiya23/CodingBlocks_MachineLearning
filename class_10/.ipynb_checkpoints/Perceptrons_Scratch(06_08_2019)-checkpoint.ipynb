# Model and Helper Functions

```python
def sigmoid(z):
  return (1.0)/(1.0 + np.exp(-z))
```

```python
z = np.array([91,2,3,4,5,867,8])
print(z.shape)
print(z)
sigmoid(z)
```

# Implement Perceptron Learning Algorithm
 - Learn Weights
 - Reduce loss
 - Make Predictions

```python
def predict(X,weights):
  '''  X = (m*n+1) matrix       weights = (n*1) vector'''
  z = np.dot(X,weights)
  predictions = sigmoid(z)
  return predictions 
```

```python
def loss(X,Y,weights):
  '''Binary Cross Entropy Loss'''
  pred = predict(X,weights)
  loss  = np.mean(-Y*np.log(pred) - (1-Y)*np.log(1-pred))
  return loss
```

```python
def update(X,Y,weights,learning_rate=0.5):
  '''Perform weight updates for 1 epoch'''
  pred = predict(X,weights)
  dw = np.dot(X.T,pred - y)
  
  m = X.shape[0]
  
  weights = weights - learning_rate*dw/float(m)
  return weights
```

```python
def training(X,Y,learning_rate = 0.5,maxEpoch = 100):
  
  
  # adding biases to the matrix
  b = np.ones((X.shape[0],1))
  X = np.hstack((b,X))
  
  # intialize weights
  weights  = np.zeros(X.shape[1])
    
  for epoch in range(maxEpoch):
    
    weights = update(X,Y,weights,learning_rate)
    
    if epoch%10 ==0:
      l = loss(X,Y,weights)
      print("Epoch %d Loss %.4f"%(epoch,l))
    
  return weights
  
```

```python
weights = training(X,y,learning_rate=0.8,maxEpoch = 1000)
```

# Perceptron Implementation 2:
- Make Predictions
- Visualize decision Surface
- Linear vs Non - Linear Data

```python
def get_pred(X_test,weights,labels = True):
  if X_test.shape[1] != weights.shape[0]:
    b = np.ones((X_test.shape[0],1))
    X_test = np.hstack((b,X_test))
    
  pred = predict(X_test,weights)
  
  if labels == False:
    return pred
  else:
    labels = np.zeros(pred.shape)
    labels[pred>=0.5] = 1
    return labels
```

```python
x1 = np.linspace(-7,-1,10)
x2 = -(weights[1]*x1+weights[0])/weights[2]
```

```python
plt.scatter(X[:,0],X[:,1],c='Y',cmap=plt.cm.Accent)
plt.plot(x1,x2,c='red')
plt.show()
```

```python
# Find Accuracy

Y_pred = get_pred(X,weights)

acc = np.sum(Y_pred == y)/X.shape[0]
print(acc)
```
